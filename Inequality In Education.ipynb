## Exploratory Data Analysis
> **Tips** Here we'll perform some analysis on the cleaned data, using statistical computation and visualization.

#### EDA Questions

1.Does inequality in education increases over the year for a country?    
2.What region has the highest number of  Low, Medium, and High for HDG ranking?  
3.What is the relationship between HDI ranking and inequality in education in 2021?  
4.What region has the highest inequality over the years

```python
# 1.Does inequality in education increases over the year for a country?
# To perform this task we need to check the relationship between Years of inequality in education and the countries.
# To visualize this, let's use a gradient map
# Let's start with 2010
ineduc[['Country','Inequality in Education (2010)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2010)'], ascending=False).head(10).style.background_gradient(subset='Inequality in Education (2010)', cmap='BuGn')
```

```python
#Here is for bottom 10 countries with least inequality in education
ineduc[['Country','Inequality in Education (2010)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2010)'], ascending=False).tail(10).style.background_gradient(subset='Inequality in Education (2010)', cmap='BuGn')
```

```python
#Here is for top 10 countries with inequality in education (2011)
ineduc[['Country','Inequality in Education (2011)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2011)'], ascending=False).head(10).style.background_gradient(subset='Inequality in Education (2011)', cmap='BuGn')
```

```python
#Here is for top 10 countries with inequality in education (2012)
ineduc[['Country','Inequality in Education (2012)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2012)'], ascending=False).head(10).style.background_gradient(subset='Inequality in Education (2012)', cmap='BuGn')
```

```python
#Here is for top 10 countries with inequality in education (2018)
ineduc[['Country','Inequality in Education (2018)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2018)'], ascending=False).head(10).style.background_gradient(subset='Inequality in Education (2018)', cmap='BuGn')
```

```python
#Here is for top 10 countries with inequality in education (2015)
ineduc[['Country','Inequality in Education (2015)','UNDP Developing Regions','Human Development Groups']].sort_values(['Inequality in Education (2015)'], ascending=False).head(10).style.background_gradient(subset='Inequality in Education (2015)', cmap='BuGn')
```

This shows the rate of inequality in education in some countries fluctuates over the years, meaning for a country like Nigeria, it decreases at a particular year and increases at some other year.
Same with Bhutan and Senegal

### Question 2
#### What region has the highest number of  Low, Medium, and High for HDG ranking? 

```python
#What region has the highest number of  Low, Medium, and High for HDG ranking? 
#To solve this we have to count number of times each region appears on low, medium, high,and ver high
region_counts = ineduc.groupby(['UNDP Developing Regions', 'Human Development Groups'])['Human Development Groups'].count().unstack(fill_value=0)
```

```python
print(region_counts)
```

```python
plt.figure(figsize=(10, 6))  # Adjust figure size as needed
region_counts.plot(kind='bar')
plt.title("Human Development Regions by Groups")
plt.xlabel("Region")
plt.ylabel("HDG")
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.grid(axis='y')  # Add gridlines for better visualization
```

This clearly shows that;
1. LAC (Latin America and the Caribbean ) regions have the highest number of High and also the lowest number
2. SSA (Sub-Saharan Africa)regions have the lowest number of high and also the highest number of low

### Question 3
#### What is the relationship between HDI ranking and inequality in education in 2021?

```python
#To check if there is a relationship between HDI rank(2021) and inequality in education (2021), let's show
#HDI rank (2021) in accending order.
ineduc[['Country','HDI Rank (2021)','Inequality in Education (2021)','UNDP Developing Regions']].sort_values(['HDI Rank (2021)'], ascending=True).head(10).style.background_gradient(subset='HDI Rank (2021)', cmap='BuGn')
```

This shows the HDI Ranking in 2021 was not based on the Inequality In education index in that same year.
Country like Argentina should have taken a higher ranking than Singapore if that was the case.

### Question 4
#### What region has the highest inequality over the years

```python
#Do carry out this task we need to add a column that sums up the entire Inequality in Education years.
ineduc.info()
```

```python
#Let's add a column
ineduc['Total Inequality In Education (2010-2021)'] = ineduc[['Inequality in Education (2010)',
                                                               'Inequality in Education (2011)',
                                                              'Inequality in Education (2012)',
                                                              'Inequality in Education (2013)',
                                                              'Inequality in Education (2014)',
                                                              'Inequality in Education (2015)',
                                                              'Inequality in Education (2016)',
                                                              'Inequality in Education (2017)',
                                                              'Inequality in Education (2018)',
                                                              'Inequality in Education (2019)',
                                                              'Inequality in Education (2020)',
                                                              'Inequality in Education (2021)']].sum(axis=1)
```

```python
#Now that we have added the column, let's check 
ineduc.head()
```

```python
#We can now check the country with the highest Inequality in Education over the years
ineduc[['Country','Total Inequality In Education (2010-2021)','UNDP Developing Regions']].sort_values(['Total Inequality In Education (2010-2021)'], ascending=False).head(10).style.background_gradient(subset='Total Inequality In Education (2010-2021)', cmap='BuGn')
```

This shows that Guinea has  the highest Total inequality in education over the years (2010-2021)
